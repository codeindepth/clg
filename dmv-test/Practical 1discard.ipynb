# Title : Analyzing Sales Data from Multiple File Formats

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_data = pd.read_csv("datasets/sales_data_sample.csv", encoding="cp1252")
```

```python
excel_data = pd.read_excel("datasets/Sample-Sales-Data.xlsx")
```

```python
json_data = pd.read_json("datasets/customers.json")
```

```python
formats = [csv_data, excel_data, json_data]
format_names = ["CSV Data", "Excel Data", "JSON Data"]

for format_data, name in zip(formats, format_names):
    print(f"____({name}) Sample Data_____")
    display(format_data.sample(5))
    
    print(f"____({name}) Data Information_____")
    format_data.info()  
    
    print(f"____({name}) Data Description_____")
    display(format_data.describe())
    
    print(f"____({name}) Check Missing Data_____")
    display(format_data.isna().sum())
    
    print(f"____({name}) Data After Dropping Missing Values_____")
    display(format_data.dropna())
    
    print(f"____({name}) Check Duplicate Data_____")
    display(format_data.duplicated().sum())
    
    print(f"____({name}) Data After Dropping Duplicates_____")
    display(format_data.drop_duplicates())
    print("\n" + "="*50 + "\n")
```

```python
unified_data = pd.concat([csv_data, excel_data], ignore_index=True)
```

```python
# display(unified_data.isna().sum())
# display(unified_data.dropna())
```

```python
total_sales = unified_data['SALES'].sum()
print("Total Sales:", total_sales)
```

```python
category_sales = unified_data.groupby('ORDERNUMBER')['SALES'].mean()
```

```python
category_counts = unified_data['SALES'].value_counts()
category_counts.plot(kind='bar')
plt.title('Product Category Distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```
