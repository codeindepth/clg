```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('datasets/Superstore.csv',encoding="cp1252")
df.head()
```

```python
relevant_variables = ['Region', 'Sales', 'Category']

region_sales = df.groupby('Region')['Sales'].sum()
print(region_sales)

plt.figure(figsize=(10, 6))
plt.bar(region_sales.index, region_sales.values)
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.title('Sales Distribution by Region')
plt.xticks(rotation=45, ha='right')
plt.show()

plt.figure(figsize=(8, 8))
plt.pie(region_sales.values, labels=region_sales.index, autopct='%1.1f%%', startangle=90)
plt.title('Sales Distribution by Region')
plt.show()

top_regions = region_sales.sort_values(ascending=False)
print("\nTop Performing Regions:")
print(top_regions)

region_category_sales = df.groupby(['Region', 'Category'])['Sales'].sum()
print("\nSales by Region and Product Category:")
region_category_sales
```
